```python
import numpy as np 
arr = np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,37,34,201])
arr
```

```python
Q1 = np.percentile(arr,25)
Q1
```

```python
Q3 = np.percentile(arr,75)
Q3
```

```python
IQR = Q3-Q1
IQR
```

```python
#upper fence and lower fence 
upper_fence = Q3 + (1.5*IQR)
upper_fence
```

```python
lower_fence = Q1 - (1.5*IQR)
lower_fence
```

```python
#remove number beyond upper and lower fence 
l=[]
for i in arr:
    if i>=lower_fence and i <=upper_fence:
        l.append(i)
arr2 = np.array(l)
        
```

```python
print(arr)
print(arr2)
```

```python
#make box plot of both 
import seaborn as sns 

sns.boxplot(x=arr)
```

```python
sns.boxplot(x = arr2)
```
