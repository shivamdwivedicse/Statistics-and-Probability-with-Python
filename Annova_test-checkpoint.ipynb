```python
import seaborn as sns 
import pandas as pd 
from scipy.stats import f_oneway 
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df = df[["age","pclass"]].dropna()
```

```python
df['pclass'].unique()
```

```python
#is there is any significiant relation between passenger class and ages 
```

```python
class_1 = df[df['pclass'] ==1]['age']
class_2 = df[df['pclass'] ==2]['age']
class_3 = df[df['pclass'] ==3]['age']

class_1
```

```python
f_stats , p_value = f_oneway(class_1,class_2,class_3)
f_stats
```

```python
p_value
```

```python
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis and there is a significant difference in atleast one passanger class")
else:
    print("Accept the null hypothesis and there is no a significant difference in atleast one passanger class")
```
