```python
import numpy as np 
from scipy.stats import norm
```

```python
sample = [172,174,168,169,171,173,175,170,169,172]
population_mean = 170
population_std = 3
sample_mean = np.mean(sample)
n=len(sample)
sample_mean
```

```python
z_score = (sample_mean - population_mean)/(population_std/np.sqrt(n))
z_score
```

```python
p_values = 2*(1-norm.cdf(abs(z_score)))
p_values
```

```python
#if p_value is less then alpha i reject null hypothesis and accept alternate hypothesis
```

```python
# if p value is greater than my myu alpha value i will accept the null hypothesis
```

```python
alpha = 0.05

if p_values < alpha:
    print("I will reject null hypothesis")
else:
    print("I will accept null hypothesis")
```
